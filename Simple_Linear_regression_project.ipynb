## Handling Missing Data

```python
missing_data = df.isnull()
missing_data.head(5)
```

### Get the number of missing_data per column

```python
for column in missing_data.columns.values.tolist():
    print(column)
    print(missing_data[column].value_counts())
    print("")
```

### How many missing_values do we have

```python
missing_values_count = df.isnull().sum()
missing_values_count
```

```python
total_cells = np.product(df.shape)
total_missing = missing_values_count.sum()

# percent of data that is missing
(total_missing/total_cells) * 100
```

**About 15.3 % of the dataset has missing_values**

## Dummy Variable

```python
dummy_variable_1 = pd.get_dummies(df["Agency"])
dummy_variable_1
```

```python
dummy_variable_1.rename(columns={"Agency":"San Francisco"}, inplace=True)
dummy_variable_1.head()
```

```python
df = pd.concat([df, dummy_variable_1], axis=1)

#drop the original data
df.drop("Agency", axis=1, inplace=True)
```

```python
df.head()
```

### Declare the column you want to use as features

```python
features = ["Id"]
```

### Specify the Prediction Target

```python
target = ["Year"]
```

### Extract features and target into separate dataframe

```python
X = df[features]
```

```python
y = df[target]
```

```python
print(X)
```

```python
print(y)
```

### Split the dataset into Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
print(y_train)
```

```python
print(y_test)
```

### Train the Simpl Linear Regression on the Training set

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

### Predict the Test result

```python
y_pred = regressor.predict(X_test)
y_pred
```

## Visualising the Training set

```python
plt.scatter(X_train, y_train, color = "red")
plt.plot(X_train, regressor.predict(X_train), color = "blue")
plt.xlabel("Id Payment")
plt.ylabel("Years of Payment")
plt.title("Id Payment vs Years of Payment ")
plt.show()
```

### Visualising the Test set

```python
plt.scatter(X_test, y_test, color = "red")
plt.plot(X_train, regressor.predict(X_train), color = "blue")
plt.xlabel("Id Payment")
plt.ylabel("Years of Payment")
plt.title("Id Payment vs Years of Payment ")
plt.show()
```

```python
print(regressor.predict([[2010]]))
```

### Evaluating Model Performace

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```
